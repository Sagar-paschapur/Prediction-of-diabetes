```python
from sklearn.impute import SimpleImputer

fill_zero=SimpleImputer(missing_values=0,strategy='mean')

x_train=fill_zero.fit_transform(x_train)
x_test=fill_zero.fit_transform(x_test)
```

## Training by using algorithm

```python
from sklearn.naive_bayes import GaussianNB

nb_model=GaussianNB() ## creating object for gaussian class
nb_model.fit(x_train,y_train.ravel()) ##training a model

## ravel is a pre defined function it is used to convert 2d to 1d
```

## 5.training data

```python
from sklearn import metrics
nb_predict_model=nb_model.predict(x_train)
print(f"predicted accuracy is:{metrics.accuracy_score(y_train,nb_predict_model)}")
```

## 6. testing data

```python
from sklearn import metrics
nb_predict_model=nb_model.predict(x_test)
print(f"predicted accuracy is:{metrics.accuracy_score(y_test,nb_predict_model)}")
```

# confusion matrix

```python
'''
compare the predicted value to oerigional value

it divided into 4 parts
1)True positive:if predicted value is TRUE origional value is TRUE
2)False positive:if predicted value is TRUE origional value is FALSE
3)True Negative:if predicted value is FALSE origional value is FALSE
4)False Negative:if predicted value is FALSE origional value is TRUE


FORMULA IS:
Accuracy=((TP+TN)/(TP+TN+FP+FN))*100
'''
```

```python
print("confusion matrix")
print(f"{metrics.confusion_matrix(y_test,nb_predict_model)}")
```

```python
print("print of classification report")
print(metrics.classification_report(y_test,nb_predict_model))
```

## Formulas

```python
Positive report
1)positive precision=tp/(tp+fp)
2)positive recall=tp/(tp+fn)
3)positive f1 score=2*((precision*recall)/(precision+recall))

Negative report
1)negative precision=tn/(tn+fn)
2)negative recall=tn/(tn+fp)
3)negative f1 score=2*((precision*recall)/(precision+recall))


macro avg precision score=(positive precision+negative precision)/2

macro avg recall score=(positive recall+negative recall)/2


weight avg=(original positive/total test data)*positive precision (recall)+(original negative/total test data)*negative precision (recall)
```
